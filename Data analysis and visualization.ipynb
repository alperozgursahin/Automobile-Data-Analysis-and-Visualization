```python
average_acceleration = data.groupby("origin")["acceleration"].mean()
```

```python
average_acceleration
```

```python
average_horsepower = data.groupby("origin")["horsepower"].mean()
average_horsepower
```

```python

origin_informations = {
    "Average Horsepower" : (average_horsepower["usa"], average_horsepower["japan"], average_horsepower["europe"]),
    "Average Acceleration" : (average_acceleration["usa"], average_acceleration["japan"] , average_acceleration["europe"]), 
}


x = np.arange(len(data["origin"].value_counts()))  # the Brand locations
width = 0.15  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in origin_informations.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

    
ax.set_xlabel('Origins')
ax.set_title('Horsepowers and Accelerations Comparision')
ax.set_xticks(x + width/2, data["origin"].value_counts().index)
ax.legend(loc='upper right', ncols=2)
ax.set_ylim(0, 130) 
plt.show()
```
