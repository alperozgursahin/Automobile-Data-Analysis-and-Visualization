```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.getmidas.com/canli-borsa/xu100-bist-100-hisseleri"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python
data = soup.find_all("table", class_ = "stock-table w-100")
```

```python
stock_table = data[0].contents[len(data[0].contents) - 2]
```

```python
stock_table = stock_table.find_all("tr")
```

```python
companies_data = []
```

```python
for stock in stock_table:
    stock_info = {}
    columns = stock.find_all("td", class_="val")
    stock_info["Stock"] = stock.find("a", class_="title stock-code").text
    stock_info["Last"] = columns[1].text.strip().replace(" ", "")
    stock_info["Buying"] = columns[2].text.strip().replace(" ", "")
    stock_info["Selling"] = columns[3].text.strip().replace(" ", "")
    stock_info["Diff"] = columns[4].text.strip().replace(" ", "")
    stock_info["Lowest"] = columns[5].text.strip().replace(" ", "")
    stock_info["Highest"] = columns[6].text.strip().replace(" ", "")
    stock_info["WAP"] = columns[7].text.strip().replace(" ", "")
    stock_info["Volume"] = columns[8].text.strip().replace(" ", "")
    companies_data.append(stock_info)
```

```python
df = pd.DataFrame(companies_data)
print(df.head())
```

```python
sorted_volume = df.sort_values(ascending=False, by= "Volume")
sorted_volume.head(10)
```
